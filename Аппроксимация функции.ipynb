```python
import numpy as np
import math
import scipy.linalg
from scipy import optimize
```

```python
def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
```

```python
a = np.array([[1, 1], [1, 15]])
b = np.array([1, 15])
b = f(b)
x = np.linalg.solve(a, b)
print(x)
```

```python
%matplotlib inline
from matplotlib import pylab as plt
c = np.arange(0, 15.1, 0.1)
y = sum([x[i]*(c**i) for i in range(len(x))])
plt.plot(c,y)
```
