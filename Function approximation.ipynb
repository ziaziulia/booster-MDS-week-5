```python
import numpy as np
import math
```

```python
def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
```

```python
a = np.array([[1, 1], [1, 15]])
b = np.array([1, 15])
b = f(b)
x = np.linalg.solve(a, b)
print(x)
```

```python
%matplotlib inline
from matplotlib import pylab as plt
plt.plot(x)
plt.show()
```

```python
a = np.array([[1, 1, 1],[1,8, 64], [1,15,225]])
b = np.array([1,8,15])
b = f(b)
x = np.linalg.solve(a, b)
print(x)
```

```python
plt.plot(x)
plt.show()
```
